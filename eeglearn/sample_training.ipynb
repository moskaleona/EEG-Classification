```python
from __future__ import print_function
from sklearn.model_selection import StratifiedKFold as SKFold
from eeg_cnn_funcs import *
import numpy as np
import os
import random
```

```python
path_to_data = '...'
path_to_results = '...'
subnums = [33,34,35]
```

```python
elNum, locs_2d = loadlocs(path_to_data)
```

```python
# Cross-validation on one subject
# Data preparation for CNN
n_folds = 4
subnum = 34
#print(data.shape)
if not os.path.isdir(path_to_results + str(subnum)):
        os.mkdir(os.path.join(path_to_results, str(subnum)))
feats = scipy.io.loadmat(os.path.join(path_to_data,'FeatureMat_timeWin' + str(subnum) +'.mat'))['features']
# CNN Mode
print('Generating images...')
# Find the average response over time windows
av_feats = reduce(lambda x, y: x + y,
                    [feats[:, i * 3 * elNum:(i + 1) * 3 * elNum] for i in range(feats.shape[1] / (3 * elNum))])
av_feats = av_feats / (feats.shape[1] / (3 * elNum))
images = gen_images(np.array(locs_2d),
                    av_feats,
                    32, normalize=False)
labels = np.squeeze(feats[:, -1])
print('\n')


fold_pairs = []
np.random.seed(108)
skf = SKFold(n_splits=n_folds, shuffle=True)

for tr, ts in skf.split(images, labels):
    random.shuffle(tr)
    random.shuffle(ts)
    print(len(tr), len(ts), np.sum(labels[tr]), np.sum(labels[ts]))
    fold_pairs.append((tr,ts))
    
print(len(labels), np.sum(labels))

     
```

```python
# Training on 1 subject (CNN)
print('Training the CNN Model for ', str(subnum), ' subject...')
for fold in range(n_folds):
    print('fold:', fold)
    train_save(images, labels, fold_pairs[fold],
               'cnn', path_to_results + str(subnum) + '/' + str(fold) + '/',
               num_epochs=100,  learning_rate=0.0003, l1_coef=1e-5, l2_coef=1e-4, dropout=0.4)
    print('\n')
```

```python
# Conv-LSTM Mode
print('Generating images for all time windows...')
images_timewin = np.array([gen_images(np.array(locs_2d),
                                        feats[:, i * (3 * elNum):(i + 1) * (3 * elNum)], 32, normalize=False) for i in
                            range(feats.shape[1] / (3 * elNum))
                            ])

print('\n')
print('Training the LSTM-CONV Model for ', str(subnum), ' subject...')
train_save(images_timewin, np.squeeze(feats[:, -1]) - 1, 'mix',
                     path_to_results + str(subnum) + '/',
                     k=4, learning_rate=0.0005, num_epochs=2)

print('Done!')

  
```

```python
#Training on all subjects separately
n_folds=3
for subnum in subnums:
    if not os.path.isdir(path_to_results + str(subnum)):
        os.mkdir(path_to_results + str(subnum))
    feats = scipy.io.loadmat(path_to_data + 'FeatureMat_timeWin' + str(subnum) +'.mat')['features']
    # CNN Mode
    print('Generating images...')
    # Find the average response over time windows
    av_feats = reduce(lambda x, y: x + y,
                        [feats[:, i * 3 * elNum:(i + 1) * 3 * elNum] for i in range(feats.shape[1] / (3 * elNum))])
    av_feats = av_feats / (feats.shape[1] / (3 * elNum))
    images = gen_images(np.array(locs_2d),
                        av_feats,
                        32, normalize=False)
    labels = np.squeeze(feats[:, -1])
    print('\n')


    fold_pairs = []
    skf = SKFold(n_splits=n_folds, shuffle=True)
    for tr, ts in skf.split(images, labels):
        random.shuffle(tr)
        random.shuffle(ts)
        fold_pairs.append((tr,ts))
    # Training 
    print('Training the CNN Model for ', str(subnum), ' subject...')
    for fold in range(n_folds):
        print('fold:', fold)
        train_save(images, labels, fold_pairs[fold],
                   'cnn', path_to_results + str(subnum) + '/' + str(fold) + '/',
                   num_epochs=100,  learning_rate=0.0003, l1_coef=1e-5, l2_coef=1e-4, dropout=0.4)

print('Done!')
```
